```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("CNI.csv")
df
```

```python
plt.xlabel('Year')
plt.ylabel('Per Capita Income $')
plt.scatter(df.Serial, df.PerCapitaIncome,color='Blue',marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Serial']],df.PerCapitaIncome)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[51]])
```
